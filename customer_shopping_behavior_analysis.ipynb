```python
df['Review Rating']=df['Review Rating'].fillna(df.groupby('Category')['Review Rating'].transform('mean'))
```

```python
df['Review Rating']
```

```python
df.columns=df.columns.str.lower().str.replace(' ', '_')
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#creating a column age_group
labels=['Young Adult','Adult','Middle Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'] = (
    df['frequency_of_purchases']
    .str.strip()      # removes spaces
    .str.title()      # Weekly, Monthly, etc.
)
```

```python
# creating a cloumn  purchase_frequency_days
frequency_mapping={'Fortnightly':14,'Weekly':7,'Monthly':30,'Quarterly':90,'Bi-Weekly':14,'Annually':365,'Every 3 Months':90}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[df['purchase_frequency_days'].isna()]['frequency_of_purchases'].unique()
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```
